```python
class Playfair:
    def __init__(self,key):
        self.dmat=[]
        self.mat=[]
        self.smat=''
        self.key=key
        
    def preprocessing(self,plaintext):
        plaintext =plaintext.lower()
        # plaintext =plaintext.replace(' ','')
        for i in range(0,len(plaintext)-1):
            if plaintext[i] == plaintext[i+1]:
                plaintext = plaintext[:i+1] +'x' +plaintext[i+1:]
        if len(plaintext)%2!=0:
            plaintext+='x'
        return plaintext
    
    def postprocessing(self,plaintext):
        # plaintext =plaintext.replace(' ','')
        for i in range(1,len(plaintext)-1):
            if plaintext[i]=='x':
                if plaintext[i-1]==plaintext[i+1]:
                    plaintext=plaintext[:i]+plaintext[i+1:]
        return plaintext
        
    def matrix(self):
        key =self.key
        data ='abcdefghiklmnopqrstuvwxyz'
        # data=sorted(set(data))
        temp= []
        for ch in key:
            if ch not in temp:
                if ch =='j':
                    if 'i' in temp:
                        continue
                    temp.append('i')
                else:
                    temp.append(ch)
        for alpha in data:
            if alpha not in temp:
                temp.append(alpha)
        m=[]
        count=0
        for i in range(5):
            t =[]
            for j in range(5):
                t.append(temp[count])
                count+=1
            m.append(t)   
            
        self.smat = "".join(temp)   
        self.mat =temp
        self.dmat= m
        
    
    def loc(self,a,b):
        mat=self.smat
        ind1=mat.index(a)
        ind2=mat.index(b)
        pos1 =[ind1//5,ind1%5]
        pos2 =[ind2//5,ind2%5]
        return pos1,pos2
    
    def subsitute(self,a,b):
        print(a,b)
        dmat=self.dmat
    # a[4,1] b[2,3]
        if a[0]==b[0]:
            a[1]=(a[1]+1)%5
            b[1]=(b[1]+1)%5
        elif a[1]==b[1]:
            a[0]=(a[0]+1)%5
            b[0]=(b[0]+1)%5
        else:
            pos_a_col =a[1]
            a[1] =b[1]
            b[1] = pos_a_col
        
        # return a,b
        return dmat[a[0]][a[1]],dmat[b[0]][b[1]]
    
    def subsituteD(self,a,b):
        print(a,b)
        dmat=self.dmat
    # a[4,1] b[2,3]
        if a[0]==b[0]:
            a[1]=(a[1]-1)%5
            b[1]=(b[1]-1)%5
        elif a[1]==b[1]:
            a[0]=(a[0]-1)%5
            b[0]=(b[0]-1)%5
        else:
            pos_a_col =a[1]
            a[1] =b[1]
            b[1] = pos_a_col
        
        # return a,b
        return dmat[a[0]][a[1]],dmat[b[0]][b[1]]
    
    def encrypt(self,plaintext):
        plaintext = self.preprocessing(plaintext)
        self.matrix()
        ciphertext=''
        for i in range(0,len(plaintext),2):
            a,b= self.loc(plaintext[i],plaintext[i+1])
            s1,s2 =self.subsitute(a,b)
            ciphertext+=s1+s2
        print(ciphertext)
        return ciphertext
    
    def decrypt(self,ciphertext):
        plaintext=''
        for i in range(0,len(ciphertext),2):
            a,b= self.loc(ciphertext[i],ciphertext[i+1])
            s1,s2 =self.subsituteD(a,b)
            plaintext+=s1+s2
        print(plaintext)
        return plaintext
```
